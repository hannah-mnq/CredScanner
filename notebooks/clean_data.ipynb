```python
import pandas as pd
import os 
import re
DATA_FOLDER = "../data/fake_news/"
```

```python
fake_df=pd.read_csv(DATA_FOLDER + "fake.csv")
true_df=pd.read_csv(DATA_FOLDER + "true.csv")

fake_df['label']='FAKE'
true_df['label']='REAL'
fake_df.head(2)
```

```python
true_df.head(2)
```

```python
full_df = pd.concat([fake_df, true_df]).sample(frac=1, random_state=42).reset_index(drop=True)
full_df.head(10)
```

```python
# cleaning text
def clean(text):
    text=str(text).lower()
    text=re.sub("http\S+|@\S+|[^a-zA-Z0-9\s]", "", text)
    return text.strip()

full_df['text'] = full_df['text'].apply(clean)
full_df.drop_duplicates(subset="text", inplace=True)
full_df.dropna(subset=["text", "label"], inplace=True)

# Step 5: Save cleaned dataset
clean_file_path = os.path.join(DATA_FOLDER, "clean_fake_news.csv")
full_df.to_csv(clean_file_path, index=False)


print("✅ Cleaned data saved at:", clean_file_path)
```
